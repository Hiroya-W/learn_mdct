# 量子化

# 線形量子化

## 2bit

```python
import numpy as np 
import matplotlib.pyplot as plt 

q = 1/2**2 
x = np.linspace(0, 1, 1000) 
y = q * np.round(x/q) 
plt.plot(x,y) 
```

## 16bit

```python
import numpy as np 
import matplotlib.pyplot as plt 

q = 1/2**16
x = np.linspace(0, 1, 1000) 
y = q * np.round(x/q) 
plt.plot(x,y) 
```

# 非線形量子化

```python
import numpy as np 
import matplotlib.pyplot as plt 

x = np.linspace(0, 1000, 1000) 
y = - np.log2(np.round(x)/2**16) 
plt.plot(x,y) 
```

```python
import numpy as np 
import matplotlib.pyplot as plt 

x = np.linspace(0, 1000, 1000) 
y = - np.log2(np.round(x)/2**16) 
y = 2**16 * 2**(-y)
plt.plot(x,y) 
```
